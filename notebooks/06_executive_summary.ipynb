## Appendix: Technical Details

### Statistical Methods Used:

1. **Multi-Armed Bandit (Thompson Sampling)**
   - Bayesian approach to adaptive experimentation
   - Beta-Bernoulli conjugate prior
   - Optimal allocation via posterior sampling

2. **Propensity Score Matching**
   - Logistic regression for propensity estimation
   - 1:1 nearest neighbor matching with caliper (0.05)
   - Balance checking via standardized mean differences
   - Average Treatment Effect on Treated (ATT) calculation

3. **ARIMA Time Series Forecasting**
   - Grid search for optimal (p,d,q) parameters
   - Augmented Dickey-Fuller stationarity test
   - ACF/PACF analysis for parameter selection
   - Model selection via AIC criterion

4. **Survival Analysis**
   - Kaplan-Meier non-parametric estimation
   - Cox Proportional Hazards regression
   - Log-rank test for group comparisons
   - Hazard ratio interpretation
